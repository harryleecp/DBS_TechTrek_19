```python
# I filled in the missing data with fillna() function. This method fills in each of the missing value with the nearest value above the missing value.
# which is also called forward-filling.
# I chose forward-filling due to the time series variables, where "AmountinStocks", "AmountinBonds" and "HasRelationshipManager"
# are time series variable. Filling in with variables from the previous data, I am able to predict the trend in the data.
# This will give a better prediction trend as compared to filling "NA" in a 100 row data.
```

```python
clients['PostalCode'] = clients['Address'].apply(lambda x: x[-6:-1])
```

```python
clients['StreetName'] = clients['Address'].apply(lambda x: x[9:-25])
```
